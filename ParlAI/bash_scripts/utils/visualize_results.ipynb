```python
from pathlib import Path
def reduce_name(name):
    # change name to only contain the configurations to group by 
    version = name[name.find('bart_')+len('bart_'):name.find("/")]
    fn = Path(name).name
    # print(version)
    # print(fn)
    config = fn[:fn.find("_uf")+len('_uf')+1]
    # print(config)
    return f"{version}/{config}"
```

```python
results["fn"] = results["fn"].apply(reduce_name)
```

```python
def normalize_df(df): 
    
    df= df.rename(columns = {
        "NEI_all_ne/hallucination_original": "no_hall_orig",
        "NEI_all_ne/hallucination_perturbed": "no_hall_pert",
        "NEI_consistency": "NEI_inconsistency",
    })
    
    def flip(col_name): 
        df[col_name] = df[col_name].apply(lambda x: 1-x) 
    flip("no_hall_orig")
    flip("no_hall_pert")
    flip("NEI_inconsistency")
    
    return df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
def plot_barplot(df, target_metrics): 
    
    datasize = "Fullshot" if "fs_False" in df["fn"].tolist()[0] else "Fewshot"
    
    version = "2.3" if "2.3" in df["fn"].tolist()[0] else "2.1"
    
    df["fn"] = df["fn"].apply(lambda x : x[:x.find("/")+1])
    
    plot = sns.catplot(
        data=df[target_metrics].melt("fn"), kind="bar",
        x="variable", y="value", hue="fn",
        ci="sd", palette="dark", alpha=.6, height=10, legend=False,
        estimator=np.median
    )

    for axes in plot.axes.flat:
        _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=90)

    plt.title(f"{datasize} MultiWOZ{version}")
    # ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
    plt.legend(loc='best', title='Config')
```
